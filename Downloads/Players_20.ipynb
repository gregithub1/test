```python
# Countries with the highest number of players
players['nationality'].value_counts()
```

```python
players['nationality'].value_counts()[0:10]
```

```python
players1 = players['nationality'].value_counts()[0:20]
```

```python
players1.plot(kind = 'bar', figsize = (18, 12))
plt.title('Nationality', fontsize = 15)
plt.ylabel('Frequency of Countries', fontsize = 13)
plt.xlabel('Countries')
plt.show()
```

```python
# Players that earn most
players_salary = players[['short_name', 'wage_eur']]
players_salary.head()
```

```python
players_salary = players_salary.sort_values('wage_eur', ascending = False)
players_salary.head()
```

```python
plt.figure(figsize = (12,9))
plt.bar(list(players_salary['short_name'])[0:6], list(players_salary['wage_eur'])[0:6])
plt.title('Top Salary Earners', fontsize = 16)
plt.ylabel('Amount earned by players', fontsize = 13)
plt.show()
```
